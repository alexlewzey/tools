```python
import plotly.express as px
import polars as pl

pl.Config.set_tbl_rows(100)
pl.Config.set_tbl_cols(100)
```

```python
df = pl.read_csv("strong.csv")
```

```python
df.select(pl.all().n_unique())
```

```python
exercises = (
    df.group_by("Exercise Name")
    .agg(pl.col("Date").n_unique())
    .sort("Date", descending=True)
    .head(30)
)
exercises["Exercise Name"].to_list()
exercises = [
    "Bench Press (Barbell)",
    "Lat Pulldown (Cable)",
    "Iso-Lateral Row (Machine)",
    "Hack Squat",
    "Triceps Pushdown (Cable - Straight Bar)",
    "Lying Leg Curl (Machine)",
    "Lateral Raise (Machine)",
    "Tib Bar",
    "Bicep Curl (Barbell)",
    "Bicep Curl (Dumbbell)",
    "Leg Press",
    "Seated Row (Cable)",
    "Pull Up (Assisted)",
    "Triceps Extension (Dumbbell)",
    "Seated Palms Up Wrist Curl (Dumbbell)",
    "Iso-Lateral Chest Press (Machine)",
    "Shoulder Press (Plate Loaded)",
    # 'Incline Bench Press (Dumbbell)',
    "Romanian Deadlift (Dumbbell)",
]
```

```python
gb = (
    df.filter(pl.col("Exercise Name").is_in(exercises))
    .filter(pl.col("Weight") < 100)
    .group_by("Exercise Name", "Date")
    .agg(pl.max("Weight"))
    .sort("Date")
)
f = px.line(gb, "Date", "Weight", color="Exercise Name", markers=True)
f
```

```python
exercise = exercises["Exercise Name"][20]
gb = (
    df.filter(pl.col("Exercise Name") == exercise)
    .filter(pl.col("Weight") < 100)
    .group_by("Date")
    .agg(pl.max("Weight"))
    .sort("Date")
)
f = px.line(gb, "Date", "Weight", title=exercise, markers=True)
f
```
